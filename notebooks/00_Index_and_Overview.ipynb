### Quick Demonstration

````python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple robotic hand simulation
hand = RoboticHand()
field = CoherenceField()

# Simulate 100 steps
desired_angles = np.ones((5, 3)) * 1.0  # Target: fully closed
angles_history = []
coherence_history = []

for step in range(100):
    hand.step(desired_angles)
    field.step(hand.get_angles())
    angles_history.append(hand.get_angles().copy())
    coherence_history.append(field.get_coherence())

# Plot results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Joint angles
angles_array = np.array(angles_history)
ax1.plot(angles_array[:, 0, 0], label='Thumb MCP')
ax1.plot(angles_array[:, 1, 0], label='Index MCP')
ax1.set_xlabel('Time Step')
ax1.set_ylabel('Angle (rad)')
ax1.set_title('Joint Angles')
ax1.legend()
ax1.grid(True)

# Coherence
ax2.plot(coherence_history)
ax2.set_xlabel('Time Step')
ax2.set_ylabel('Coherence')
ax2.set_title('System Coherence')
ax2.grid(True)

plt.tight_layout()
plt.show()

print(f"‚úÖ Final coherence: {coherence_history[-1]:.6f}")
print(f"‚úÖ Average angle: {np.mean(angles_array[-1]):.6f} rad")
````
