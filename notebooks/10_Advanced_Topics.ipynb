## 1. Quantum Coherence Field

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import numpy as np
import matplotlib.pyplot as plt
from primal_logic.field import CoherenceField

# Create field with various theta values
thetas = [0.4, 0.8, 1.2, 1.6]
results = {}

for theta in thetas:
    field = CoherenceField(theta=theta)
    coherences = []
    
    for step in range(200):
        # Simulate varying joint states
        states = np.random.rand(5, 3) * 0.1 + 0.5
        field.step(states)
        coherences.append(field.get_coherence())
    
    results[theta] = coherences

# Plot
plt.figure(figsize=(12, 6))
for theta, coherences in results.items():
    plt.plot(coherences, label=f'θ = {theta}', linewidth=2)

plt.xlabel('Time Step', fontsize=12)
plt.ylabel('Coherence', fontsize=12)
plt.title('Coherence Field Evolution for Different θ', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

for theta, coherences in results.items():
    print(f'θ = {theta}: mean coherence = {np.mean(coherences):.6f}')
```
